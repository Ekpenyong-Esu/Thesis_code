```python
import matplotlib.pyplot as plt
import numpy as np

# Define the file paths for realtime data
file_paths_realtime = [
    r"C:\Users\mahon\Desktop\Videolatency\Node1\latency1\processing_latency.txt",
    r"C:\Users\mahon\Desktop\Videolatency\Node2\pythonProject\processing_latency.txt",
    r"C:\Users\mahon\Desktop\Videolatency\Node2\pythonProject\communication_latency.txt",
    r"C:\Users\mahon\Desktop\Videolatency\Node3\pythonProject1\processing_latency.txt",
    r"C:\Users\mahon\Desktop\Videolatency\Node3\pythonProject1\communication_latency.txt",
]

# Define the legend labels for realtime data
legend_labels_realtime = ['RT_PROC1', 'RT_PROC2', 'RT_COM2_TSN', 'RT_PROC3', 'RT_COM_4G']

# Function to calculate average, maximum, and minimum latency from a file
def calculate_latency_stats(file_path, num_lines=1000):
    with open(file_path, 'r') as file:
        latencies = [float(line.strip()) for line in file.readlines()[:num_lines]]
    return np.mean(latencies), np.max(latencies), np.min(latencies)

# Calculate latencies for each file in realtime data
stats_realtime = [calculate_latency_stats(path) for path in file_paths_realtime]

# Extract Average, Maximum, and Minimum latencies
average_latencies_realtime, max_latencies_realtime, min_latencies_realtime = zip(*stats_realtime)

# Create a grouped bar chart with different colors for each legend label
bar_width = 0.35  # Reduced width for better separation

colors_average = plt.cm.viridis(np.linspace(0, 1, len(legend_labels_realtime)))
colors_maximum = plt.cm.plasma(np.linspace(0, 1, len(legend_labels_realtime)))
colors_minimum = plt.cm.inferno(np.linspace(0, 1, len(legend_labels_realtime)))

plt.figure(figsize=(12, 10), dpi=150)

# Plot bars for realtime data - Average Latency
bottoms_realtime = np.zeros(len(file_paths_realtime))
for i in range(len(legend_labels_realtime)):
    plt.bar('Average RealTime', average_latencies_realtime[i], bar_width, bottom=bottoms_realtime, label=f'{legend_labels_realtime[i]} (Average)', color=colors_average[i])
    bottoms_realtime += average_latencies_realtime[i]

# Plot bars for realtime data - Maximum Latency
bottoms_realtime = np.zeros(len(file_paths_realtime))
for i in range(len(legend_labels_realtime)):
    plt.bar('Maximum RealTime', max_latencies_realtime[i], bar_width, bottom=bottoms_realtime, label=f'{legend_labels_realtime[i]} (Maximum)', color=colors_maximum[i])
    bottoms_realtime += max_latencies_realtime[i]

# Plot bars for realtime data - Minimum Latency
bottoms_realtime = np.zeros(len(file_paths_realtime))
for i in range(len(legend_labels_realtime)):
    plt.bar('Minimum RealTime', min_latencies_realtime[i], bar_width, bottom=bottoms_realtime, label=f'{legend_labels_realtime[i]} (Minimum)', color=colors_minimum[i])
    bottoms_realtime += min_latencies_realtime[i]

plt.ylabel('Latency (ms)')
plt.title('RealTime Latency Range - RealTime')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=True, ncol=5)

plt.show()
```
