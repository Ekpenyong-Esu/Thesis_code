```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read data from the first file
file_path1 = r'C:\Users\mahon\Desktop\Videolatency\Node1\latency1\processing_latency.txt'
with open(file_path1, 'r') as file:
    node1_processing_latency = [float(line.strip()) for line in file.readlines()]

# Read data from the second file
file_path2 = r'C:\Users\mahon\Desktop\Videolatency\nonRealtime\node1\pythonProject2\processing_latency1.txt'
with open(file_path2, 'r') as file:
    node1_non_realtime_latency = [float(line.strip()) for line in file.readlines()]

# Create a DataFrame from the data
data = {
    'Node': ['Node1 With RT_PATCH'] * len(node1_processing_latency) + ['Node1 With Non RT_PATCH'] * len(node1_non_realtime_latency),
    'Latency': node1_processing_latency + node1_non_realtime_latency
}

df = pd.DataFrame(data)

# Set style
sns.set(style="whitegrid")

# Plotting box plot with more features
plt.figure(figsize=(12, 8))
sns.boxplot(x='Node', y='Latency', data=df, palette="Set3", notch=True)
plt.title('Node1 Processing Latency with RT_PATCH and Non RT_PATCH', fontsize=16)
plt.xlabel('Node One', fontsize=14)
plt.ylabel('Latency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```
